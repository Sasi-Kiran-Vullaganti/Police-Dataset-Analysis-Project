**GroupBy**


*   syntax : dataset.groupby("col1").col2



```python
police.groupby("driver_gender").search_conducted.sum()
# this returns the total number of records by grouping the driver_gender and search_conducted and returns the when the
# search conducted is true.
```

```python
police.search_conducted.value_counts()
# value_counts() returns the count of each value in the column.
```

**Mapping** - mapping the old values to new values in a column


*   syntax : dataset['column'].map({'a':b,'c':d})



```python
police['stop_duration']=police['stop_duration'].map({'0-15 Min':7,'16-30 Min':24,'30+ Min':45})
# mapping the old values to new values in a column
```

```python
police.head(4)
```

```python
police['stop_duration'].value_counts()
```

**Mean** - used to calculate mean of the entire column in the dataset


*   Syntax : dataset['column'].mean()



```python
police['stop_duration'].mean()
```
